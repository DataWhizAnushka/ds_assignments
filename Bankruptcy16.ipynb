```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import zipfile
```

```python
# Loading the Excel file
data = pd.read_excel('Bankruptcy.xlsx')
```

```python
# Overview of the dataset
data.shape
```

```python
# Summary statistics
data.describe()
```

```python
# Checking for missing values
print(data.isnull().sum())
```

```python
# Class distribution
plt.figure(figsize=(2, 4))
sns.countplot(x='class', data=data, palette='viridis')
plt.title("Target Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
# 5. Correlation matrix
plt.figure(figsize=(6, 3))
# Calculate correlation for numeric columns only
sns.heatmap(data.select_dtypes(include=np.number).corr(),
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
#  Pairplot for relationships between numerical features
sns.pairplot(data, hue='class', diag_kind='kde', palette='husl')
plt.show()
```
