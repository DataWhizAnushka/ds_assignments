**Standardization of Numerical Variables**

‚óè	Objective: To scale numerical variables for uniformity, improving the dataset‚Äôs suitability for analytical models.

```python
## 	Explain the concept of standardization (z-score normalization).
```

```python
Standardization, also known as z-score normalization, is a data preprocessing technique used to transform numerical data into a standard format with a mean of 0 and a standard deviation of 1. This process helps in making data more interpretable and suitable for certain machine learning algorithms that assume normally distributed features.
```

```python
## 	Standardize the numerical columns using the formula: z=x-mu/sigma
```

```python
def standardize_column(column):
    mean = column.mean()
    std_dev = column.std()
    standardized_values = (column - mean) / std_dev
    return standardized_values
```

```python
for column in df.select_dtypes(include='int'):
    df[column] = standardize_column(df[column])
    print(df)
```

**Estimation And Confidence Intervals**



a. Build 99% Confidence Interval Using Sample Standard Deviation

```python
import numpy as np
from scipy.stats import t

data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
n = len(data)
df = n - 1
confidence_level = 0.99

t_score = t.ppf((1 + confidence_level) / 2, df)
margin_of_error = t_score * (std_dev / np.sqrt(n))

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Mean: {mean}")
print(f"Standard Deviation: {std_dev}")
print(f"Sample Size: {n}")
print(f"Degrees of Freedom: {df}")
print(f"t-score: {t_score}")
print(f"Margin of Error: {margin_of_error}")
print(f"99% Confidence Interval: ({lower_bound}, {upper_bound})")
```

b. Build 99% Confidence Interval Using Known Population Standard Deviation

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
population_std_dev = 0.2
mean = np.mean(data)
n = len(data)
z_score = 2.576
margin_of_error = z_score * (population_std_dev / np.sqrt(n))

lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Mean: {mean}")
print(f"Sample Size: {n}")
print(f"Population Standard Deviation: {population_std_dev}")
print(f"Z-score: {z_score}")
print(f"Margin of Error: {margin_of_error}")
print(f"99% Confidence Interval: ({lower_bound}, {upper_bound})")
```
