### **DATA PREPROCESSING AND FEATURE ENGINEERING IN MACHINE LEARNING**

DATA PREPROCESSING:


```python
import pandas as pd
df =  pd.read_csv("adults_with_headers.csv")                ##exploring the dataset
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()                 ## exploring the datatype and missing values
```

```python
# Replace missing values indicated by '?' with NaN
df.replace('?', pd.NA, inplace=True)
```

```python
## Apply Scaling Techniques to Numerical Features
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# List of numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
# Standard Scaling
scaler_standard = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[numerical_features] = scaler_standard.fit_transform(df[numerical_features])
print(df_standard_scaled.head())
```

```python
# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_minmax_scaled = df.copy()
df_minmax_scaled[numerical_features] = scaler_minmax.fit_transform(df[numerical_features])
print(df_minmax_scaled.head())
```

Discussion on Scaling Techniques:

- Standard Scaling: StandardScaler standardizes features by removing the mean and scaling to unit variance. It's preferred when the data follows a Gaussian distribution.


- Min-Max Scaling: MinMaxScaler scales the data to a fixed range, typically [0, 1]. It's useful when the data does not follow a Gaussian distribution or when the model does not assume any specific distribution.

ENCODING TECHNIQUES:

```python
categorical_features = df.select_dtypes(include=['object']).columns
# Identifying features with less than 5 unique categories
features_less_than_5 = [col for col in categorical_features if df[col].nunique() < 5]

# One-Hot Encoding
df_onehot_encoded = pd.get_dummies(df, columns=features_less_than_5)
print(df_onehot_encoded.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Identifying features with more than 5 unique categories
features_more_than_5 = [col for col in categorical_features if df[col].nunique() >= 5]

# Label Encoding
df_label_encoded = df.copy()
for col in features_more_than_5:
    le = LabelEncoder()
    df_label_encoded[col] = le.fit_transform(df_label_encoded[col])

print(df_label_encoded.head())
```

Discussion on Encoding Techniques:

- One-Hot Encoding: Useful for categorical features with a small number of categories. It avoids ordinality issues but increases dimensionality.


- Label Encoding: Suitable for categorical features with many categories. It preserves ordinality but can introduce unintended relationships between categories.

Feature Engineering:

```python
# Example new features
df['capital_net'] = df['capital_gain'] - df['capital_loss']
df['age_per_education'] = df['age'] / df['education_num']

print(df.head())

```

Rationale:

- capital_net: The net capital gain or loss can be more indicative of financial stability.

- age_per_education: This feature might reveal the education level adjusted for age, which could be relevant for income prediction.

```python
## Apply Transformation to Skewed Numerical Feature
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Checking the distribution of 'capital-gain'
sns.histplot(df['capital_gain'], kde=True)
plt.show()

# Applying log transformation to 'capital-gain'
df['capital-gain-log'] = np.log1p(df['capital_gain'])

# Checking the distribution after transformation
sns.histplot(df['capital-gain-log'], kde=True)
plt.show()
```

Justification: The 'capital-gain' feature is heavily skewed. Log transformation helps in normalizing the distribution, making it more suitable for machine learning models.

FEATURE SELECTION :

```python
from sklearn.ensemble import IsolationForest

# Isolation Forest
iso = IsolationForest(contamination=0.1)
outliers = iso.fit_predict(df[numerical_features])

# Adding the outlier results to the data
df['outliers'] = outliers

# Removing outliers
df_cleaned = df[df['outliers'] != -1]

print(df_cleaned.head())
```

Discussion: Outliers can distort the training process of machine learning models, leading to poorer performance. Removing them helps in stabilizing the model.

```python
## Apply PPS to Find Relationships Between
!pip install ppscore
import ppscore as pps

# Calculating PPS matrix
pps_matrix = pps.matrix(df_cleaned)
print(pps_matrix.head())

# Correlation matrix
corr_matrix = df_cleaned.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()

print(corr_matrix)
```

**Discussion**: PPS helps in understanding the predictive power of one feature to predict another, capturing non-linear relationships which the correlation matrix might miss. Comparing both can provide a comprehensive understanding of feature interactions.
